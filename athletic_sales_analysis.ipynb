```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
csv_path = pd.read_csv("Resources/athletic_sales_2020.csv")
csv_path2 = pd.read_csv("Resources/athletic_sales_2021.csv")
#movies_df = pd.read_csv(csv_path)
as_2020 = csv_path
as_2021 = csv_path2
```

```python
# Display the 2020 sales DataFrame
as_2020.head()
```

```python
# Display the 2021 sales DataFrame
as_2021.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
as_2020.dtypes
```

```python
# Check the 2021 sales data types.
as_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([as_2020, as_2021], axis = 0)
combined_df = combined_df.reset_index(drop = True)
combined_df.head()
```

```python
# Check if any values are null.
combined_df.isnull()
```

```python
# Check the data type of each column
combined_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
copy_df = combined_df.copy()
copy_df['invoice_date'] = pd.to_datetime(copy_df['invoice_date'])
print(copy_df)
#print(copy_df.dtypes)
```

```python
# Confirm that the "invoice_date" data type has been changed.
copy_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
total_products_sold = copy_df.groupby(["region", "state", "city"])["units_sold"].agg(Total_Products_Sold=("sum"))

# Show the top 5 results.
total_products_sold.sort_values(by = "Total_Products_Sold", ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold = pd.pivot_table(copy_df, values = "units_sold", index = ["region", "state", "city"], aggfunc = "sum")

# Rename the "units_sold" column to "Total_Products_Sold"
total_products_sold = total_products_sold.rename(columns = {"units_sold" : "Total_Products_Sold"})

# Show the top 5 results.
total_products_sold.sort_values(by = "Total_Products_Sold", ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total ales"
total_sales = copy_df.groupby(["region", "state", "city"])["total_sales"].agg(Total_Sales=("sum"))


# Show the top 5 results.
total_sales.sort_values(by = "Total_Sales", ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_pivot = pd.pivot_table(copy_df, values = "total_sales", index = ["region", "state", "city"], aggfunc = "sum")
total_sales_pivot

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_pivot = total_sales_pivot.rename(columns = {"total_sales" : "Total Sales"})

# Show the top 5 results.
total_sales_pivot.sort_values(by = "Total Sales", ascending=False).head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_retailer_sales = copy_df.groupby(["retailer", "region", "state", "city"])["total_sales"].agg(Total_Sales=("sum"))


# Show the top 5 results.
total_retailer_sales.sort_values(by = "Total_Sales", ascending=False).head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
invdatetotsales = pd.pivot_table(copy_df, values = "total_sales", index = ["invoice_date"])

# Optional: Rename the "total_sales" column to "Total Sales"
invdatetotsales_ = invdatetotsales.rename(columns = {"total_sales" : "Total Sales"})

# Show the table.
invdatetotsales_
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_totals = invdatetotsales_.resample('D').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_totals.sort_values(by = "Total Sales", ascending=False).head(5)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
daily_totals = invdatetotsales_.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
daily_totals.sort_values(by = "Total Sales", ascending=False).head(5)
```
